```python
import pandas as pd
```

```python
def get_coords(index):
    coords = pd.DataFrame(index=index)
    coords['x'] = index.str.split('x').str.get(0).map(float)
    coords['y'] = index.str.split('x').str.get(1).map(float)
    return coords
```

```python
df = pd.read_table('data/Layer2_BC_count_matrix-1.tsv', index_col=0)
df = df.T[df.sum(0) >= 3].T  # Filter practically unobserved genes
sample_info = get_coords(df.index)

X = sample_info[['x', 'y']]
dfm = np.log10(df + 1)
```

```python
plt.figure(figsize=(6, 4), dpi=80)

plt.xscale('log')
plt.yscale('log')
plt.scatter(results['max_s2_t_hat'], results['qval'], alpha=0.1, rasterized=True);
plt.xlim(None, 1e1)

tmp = results.query('g in ["IGFBP5", "SPARC", "VIM", "FN1", "POSTN", "MUCL1", "PIP", "SCGB2A2",' \
                        '"GAS6", "KRT17", "PEG10", "AREG", "MNP14", "DCN", "FOXA1"]')
plt.scatter(tmp['max_s2_t_hat'], tmp['qval'], s=10, zorder=10);
texts = []
for i in tmp.index.union(top_per_ls.index):
    if results.loc[i, 'max_s2_t_hat'] == 0.:
        continue

    texts.append(plt.text(results.loc[i, 'max_s2_t_hat'], results.loc[i, 'qval'], results.loc[i, 'g']))
    
print(adjust_text(texts, arrowprops=dict(arrowstyle='-'),
                         force_points=0.1,
                         force_text=0.1,
                         force_objects=0.1,
                         zorder=20,
                         x=results['max_s2_t_hat'],
                         y=results['qval']))

plt.scatter(top_per_ls['max_s2_t_hat'], top_per_ls['qval'], s=10);
    
plt.xlabel('$\hat{\sigma}^2_t$');
plt.ylabel('Adj. P-value');
plt.gca().invert_yaxis()

plt.savefig('sigma_qval_BC.pdf', bbox_inches='tight');
```
