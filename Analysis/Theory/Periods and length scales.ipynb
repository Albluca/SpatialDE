```python
N = 8
ps = np.logspace(np.log10(1.), np.log10(30), N)
```

```python
figsize(13, 2)
for i, p in enumerate(ps):
    K = SpatialDE.base.cosine_kernel(X, p)
    plt.subplot(1, N, i +1)
    
    for j in range(3):
        mu = np.random.uniform(1, 5)
        Y = np.random.multivariate_normal(mu * I, K)
        plt.plot(X, Y, c='w', lw=3);
        plt.plot(X, Y, c='C1');
        
    plt.title('p = {0:.1f}'.format(p))
    plt.xlabel('x (Âµm)')
    plt.ylabel('Expression')
        
plt.tight_layout()
plt.savefig('periodic_examples.pdf', bbox_inches='tight')
```

```python
figsize(13, 2)
for i, p in enumerate(ps):
    K = SpatialDE.base.SE_kernel(X, p)
    plt.subplot(1, N, i +1)
    
    for j in range(3):
        mu = np.random.uniform(1, 5)
        Y = np.random.multivariate_normal(mu * I, K)
        plt.plot(X, Y, c='w', lw=3);
        plt.plot(X, Y, c='C0');
        
    plt.title('l = {0:.1f}'.format(p))
    plt.xlabel('x (Âµm)')
    plt.ylabel('Expression')
        
plt.tight_layout()
plt.savefig('rbf_examples.pdf', bbox_inches='tight')
```
