```python
%pylab inline
import pandas as pd
```

```python
from glob import glob
```

```python
results = {}
for i, fname in enumerate(glob('bootstrap_cb_result_*.csv')):
    results[i] = pd.read_csv(fname, index_col=0).sort_values('g')  # Sorting is important
```

```python
h = pd.concat(results)
```
