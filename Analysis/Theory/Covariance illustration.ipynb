```python
x = np.linspace(-1.2, 1.2, 50)
y = np.linspace(-1.2, 1.2, 50)

x, y = np.meshgrid(x, y)

x_ = x.flatten()
y_ = y.flatten()

X = np.vstack((x_, y_)).T

mask = (y_ > np.ma.sqrt(1 - ((x_ + 0.5) * 2) ** 2) / 2).filled(False)
mask ^= (y_ > -np.ma.sqrt(1 - ((x_ - 0.5) * 2) ** 2) / 2).filled(False)
mask &= (y_ < np.ma.sqrt(1 - x_ ** 2)).filled(False)
```

```python
K = np.zeros((X.shape[0],) * 2)
I = np.eye(K.shape[0])

Y = np.random.multivariate_normal(0 * X[:, 0] + 2.5, 2. * (0 * K + 1. * I))
```

```python
figsize(5, 5)
plt.scatter(*X[mask].T, c=Y[mask], s=75, marker='s');
plt.axis('equal');
plt.title('H_0');

plt.xticks([])
plt.yticks([])

ax = plt.gca()

plt.hlines(1.25, -0.75, -0.75 + 0.5, lw=5)

for spine in ax.spines.values():
    spine.set_edgecolor('k')
    spine.set_lw(3)
    
plt.tight_layout();
```

```python
l = 0.1
K = SpatialDE.base.SE_kernel(X, l)
I = np.eye(K.shape[0])

Y = np.random.multivariate_normal(0 * X[:, 0] + 2.5, 2. * (1.00 * K + 1e-8 * I))
```

```python
figsize(5, 5)
plt.scatter(*X[mask].T, c=Y[mask], s=75, marker='s');
plt.axis('equal');
plt.title('H_s');

plt.xticks([])
plt.yticks([])

ax = plt.gca()

plt.hlines(1.25, -0.75, -0.75 + 0.5, lw=5)
plt.hlines(1.19, -0.75, -0.75 + l, lw=5, color='C0')

for spine in ax.spines.values():
    spine.set_edgecolor('C0')
    spine.set_lw(3)
    
plt.tight_layout();
```
